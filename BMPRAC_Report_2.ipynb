#Background & Rationale:
Thyroid hormones—TSH, Free T4, and Free T3—play critical roles in maintaining metabolic balance, mood regulation, and overall physiological homeostasis. Disruptions in these hormones have been associated with cardiovascular disease, fatigue, depression, and metabolic syndrome (Chaker et al., 2017). While thyroid dysfunction is well documented in clinical settings, the influence of modifiable factors such as sleep duration, body mass index (BMI), depression, and vitamin D status on thyroid hormone levels at the population level remains insufficiently explored.

Each of these variables is biologically plausible as a determinant of thyroid function:

1. Sleep affects circadian regulation of the hypothalamic-pituitary-thyroid axis.

2. BMI may alter hormone levels through metabolic and inflammatory pathways.

3. Depression involves neuroendocrine dysregulation that may influence thyroid activity.

4. Vitamin D, a secosteroid hormone, has emerging links to autoimmune thyroid disease and may modulate endocrine function more broadly.

This study uses the National Health and Nutrition Examination Survey (NHANES), which provides rich, representative data on U.S. adults, to investigate these associations comprehensively. The aim is to inform preventive health strategies that go beyond clinical thresholds and consider behavioral and physiological contributors to thyroid regulation in the general population.

# Justification of Novelty with Short Literature Review:
Existing literature suggests individual associations between thyroid function and lifestyle or psychological factors:

- Sleep: Disrupted or short sleep has been linked to elevated TSH (Kim et al., 2019).

- BMI: Positively associated with TSH and inversely with Free T4 (Ruhla et al., 2010).

- Depression: Associated with subclinical hypothyroidism and altered T3 levels (Hage & Azar, 2012).

- Vitamin D: Some studies suggest a potential inverse relationship with TSH and links to autoimmune thyroiditis, though findings are mixed (Chailurkit et al., 2013).

However, existing studies are typically limited in one or more of the following ways:

- Small, clinical, or non-representative samples

- Focus on only one predictor or one thyroid hormone

- Limited attention to interactive or combined effects

```python
 import pandas as pd

# Define URLs for 3 cycles of THYROD, DEMO, BMX, SLQ, DPQ and VID data
data_urls = {
    "2007-2008": {
        "THYROD": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/THYROD_E.xpt",
        "DEMO": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DEMO_E.xpt",
        "BMX": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/BMX_E.xpt",
        "SLQ": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/SLQ_E.xpt",
        "DPQ": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DPQ_E.xpt",
        "VID": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/VID_E.xpt"
    },
    "2009-2010": {
        "THYROD": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/THYROD_F.xpt",
        "DEMO": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/DEMO_F.xpt",
        "BMX": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/BMX_F.xpt",
        "SLQ": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/SLQ_F.xpt",
        "DPQ": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/DPQ_F.xpt",
        "VID": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/VID_F.xpt"
    },
    "2011-2012": {
        "THYROD": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/THYROD_G.xpt",
        "DEMO": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/DEMO_G.xpt",
        "BMX": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/BMX_G.xpt",
        "SLQ": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/SLQ_G.xpt",
        "DPQ": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/DPQ_G.xpt",
        "VID": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/VID_G.xpt"
    }
}

# Store merged data from each cycle
all_merged = []

for cycle, files in data_urls.items():
    try:
        # Load each dataset
        thy = pd.read_sas(files["THYROD"])
        demo = pd.read_sas(files["DEMO"])
        bmx = pd.read_sas(files["BMX"])
        slq = pd.read_sas(files["SLQ"])
        dpq = pd.read_sas(files["DPQ"])
        vid = pd.read_sas(files["VID"])


        # Merge all on SEQN
        merged = pd.merge(thy, demo, on="SEQN", how="inner")
        merged = pd.merge(merged, bmx, on="SEQN", how="inner")
        merged = pd.merge(merged, slq, on="SEQN", how="inner")
        merged = pd.merge(merged, dpq, on="SEQN", how="inner")
        merged = pd.merge(merged, vid, on="SEQN", how="inner")

        merged["Cycle"] = cycle

        all_merged.append(merged)
        print(f"✅ Merged: {cycle}")
    except Exception as e:
        print(f"❌ Error loading {cycle}: {e}")

# Combine all into one DataFrame
df_thyroid = pd.concat(all_merged, ignore_index=True)
print("✅ Final shape:", df_thyroid.shape)

# Preview the first few rows
df_thyroid.head()
```

```python
# Classify TSH
def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.45:
        return 'Low (Hyperthyroid)'
    elif val > 4.5:
        return 'High (Hypothyroid)'
    else:
        return 'Normal'

# Classify Free T4
def classify_t4(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.8:
        return 'Low'
    elif val > 1.8:
        return 'High'
    else:
        return 'Normal'

# Classify Free T3
def classify_t3(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return 'Low'
    elif val > 4.4:
        return 'High'
    else:
        return 'Normal'

# Create new columns
df_subset['TSH_category']  = df_subset['LBXTSH1'].apply(classify_tsh)
df_subset['FT4_category']  = df_subset['LBXT4F'].apply(classify_t4)
df_subset['FT3_category']  = df_subset['LBXT3F'].apply(classify_t3)
```

Purpose:
This code classifies thyroid hormone lab values (TSH, Free T4, Free T3) into clinical categories like “Normal”, “Low”, or “High”.

How It Works:

**TSH (Thyroid Stimulating Hormone):**

If the value is:

1) Less than 0.45 → categorized as "Low (Hyperthyroid)"

2) Greater than 4.5 → "High (Hypothyroid)"

3) Otherwise → "Normal"

**Free T4 (Thyroxine):**

Categorized as:

1) "Low" if below 0.8

2) "High" if above 1.8

3) Otherwise → "Normal"

**Free T3 (Triiodothyronine):**

Categorized as:

1) "Low" if below 2.0

2) "High" if above 4.4

3) Otherwise → "Normal"



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Categorize TSH
def categorize_tsh(value):
    if pd.isna(value):
        return np.nan
    elif value < 0.45:
        return "Low (Hyperthyroid)"
    elif value > 4.5:
        return "High (Hypothyroid)"
    else:
        return "Normal"

# Create the column
df_thyroid['TSH_category'] = df_thyroid['LBXTSH1'].apply(categorize_tsh)

# Proceed with plotting
df = df_thyroid.copy()
categorical_vars = ['Gender', 'Ethnicity']  # <-- Ensure these columns exist

# Loop through each categorical variable
for var in categorical_vars:
    proportion_df = (
        df.groupby([var, 'TSH_category'])
        .size()
        .reset_index(name='count')
    )

    total_counts = proportion_df.groupby(var)['count'].transform('sum')
    proportion_df['percent'] = 100 * proportion_df['count'] / total_counts

    plt.figure(figsize=(10, 6))
    ax = sns.barplot(
        data=proportion_df,
        x=var,
        y='percent',
        hue='TSH_category',
        palette='Set2'
    )

    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='edge', fontsize=9, padding=2)

    plt.title(f'Percentage of Thyroid Categories by {var}', pad=20)
    plt.xlabel(var)
    plt.ylabel('Percentage')
    plt.xticks(rotation=45, ha='right')

    plt.legend(
        title='TSH Category',
        loc='center left',
        bbox_to_anchor=(1.02, 0.5),
        frameon=False
    )

    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout(rect=[0, 0, 0.85, 1])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Step 1: Define URLs for 3 cycles of MCQ data
mcq_urls = {
    "2007-08": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/MCQ_E.xpt",
    "2009-10": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/MCQ_F.xpt",
    "2011-12": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/MCQ_G.xpt"
}

# Store MCQ data from each cycle
all_mcq = []

for cycle, url in mcq_urls.items():
    try:
        df_mcq = pd.read_sas(url, format='xport')
        df_mcq["Cycle"] = cycle
        all_mcq.append(df_mcq)
        print(f"✅ Loaded: {cycle}")
    except Exception as e:
        print(f"❌ Error loading {cycle}: {e}")

# Step 2: Concatenate all into one DataFrame
df_merged = pd.concat(all_mcq, ignore_index=True)

# Step 3: Filter only rows with valid answers to "ever told had thyroid problem?" (1=Yes, 2=No)
df_thyroid_dx = df_merged[df_merged['MCQ160K'].isin([1, 2])].copy()

# Step 4: Calculate percentage diagnosed (Yes = 1)
trend = (
    df_thyroid_dx.groupby('Cycle')['MCQ160K']
    .apply(lambda x: (x == 1).mean() * 100)
    .reset_index(name='Diagnosed (%)')
)

# Convert Cycle to numeric index for regression (optional: use cycle mid-year)
cycle_numeric = [2008, 2010, 2012]  # approximate mid-years
trend['Cycle_num'] = cycle_numeric

# Step 5: Calculate slope and R value
x = trend['Cycle_num']
y = trend['Diagnosed (%)']

# Calculate slope (m) and intercept (b)
m, b = np.polyfit(x, y, 1)

# Calculate Pearson's R
r, p_value = pearsonr(x, y)

# Step 6: Plot with slope and R annotated
plt.figure(figsize=(8, 5))

# Plot actual data points
plt.plot(x, y, marker='o', linestyle='-', label='Diagnosis %')

# Plot regression line (aligned properly with numeric x)
x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', linestyle='--', label='Trend line')

# Set custom x-ticks and labels
plt.xticks(x, trend['Cycle'])  # Use mid-year x as ticks, but label them as "2007–08" etc.

# Plot formatting
plt.title('Trend in Thyroid Problem Diagnosis (2007–2012)')
plt.xlabel('NHANES Cycle')
plt.ylabel('Diagnosed with Thyroid Problem (%)')
plt.ylim(0, max(y) + 5)
plt.grid(True)

# Annotate slope and R on plot
plt.text(0.05, 0.95, f"Slope = {m:.3f}\nR = {r:.3f}", transform=plt.gca().transAxes,
         verticalalignment='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pingouin as pg
import scikit_posthocs as sp

# Function to run Welch's ANOVA + Games–Howell + log-transformed Plot
def run_anova_and_plot(df, outcome, group):
    print(f"\n===== {outcome} by {group} =====")

    # Welch's ANOVA on raw values
    try:
        welch = pg.welch_anova(dv=outcome, between=group, data=df)
        print("\nWelch's ANOVA:\n", welch)
    except Exception as e:
        print(f"\nError running Welch's ANOVA: {e}")

    # Games–Howell post-hoc on raw values
    try:
        gh = sp.posthoc_gameshowell(df, val_col=outcome, group_col=group)
        print("\nGames–Howell post-hoc:\n", gh)
    except Exception as e:
        print(f"\nError running Games-Howell post-hoc: {e}")

    # Plot log-transformed values for visualization
    plot_data = df.copy()
    plot_data[outcome] = np.log1p(plot_data[outcome])  # log(x+1) to handle zeros

    plt.figure(figsize=(7,5))
    sns.boxplot(x=group, y=outcome, data=plot_data, palette="Set3", showmeans=True,
                meanprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"black"})
    plt.title(f"log-transformed {outcome} by {group}")
    plt.xlabel(group)
    plt.ylabel(f"log({outcome} + 1)")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()

# Outcomes and groups
outcomes = ["LBXTSH1", "LBXT4F", "LBXT3F"]
groups = ["BMI_cat", "Sleep_cat", "Depression_cat", "VitD_cat"]

# Loop over all combinations
for g in groups:
    for o in outcomes:
        if df_thyroid[g].isnull().any():
            print(f"\nSkipping analysis for {o} by {g} due to missing values in grouping variable.")
            continue
        run_anova_and_plot(df_thyroid, o, g)

```

Many clinical and biological variables, such as hormone levels, are positively skewed — a few individuals may have very high values that stretch the scale. This can make it difficult to visually compare group differences in standard plots.

Importantly, this transformation was applied only for visualization. **All statistical analyses** were conducted on the **original, untransformed data** to maintain interpretability and accuracy.

AI-guided regression syntax also helped us decide to include interaction terms between BMI and sleep duration, which were not initially in our plan, but ended up revealing important combined effects on thyroid hormone levels that enhanced our understanding of the data.

# Reflection:

**Learning from AI Use**

 Using AI throughout this project significantly enhanced our data science workflow. We discovered how AI can automate tedious and error-prone tasks such as renaming variables consistently across multiple datasets, merging complex files, and transforming data formats. This saved us valuable time and reduced the risk of manual errors. AI also introduced us to alternative analytical techniques and visualization styles—such as non-parametric normality tests and QQ plots—which deepened our understanding of appropriate methods for exploring population health data.

The AI’s ability to generate clear, ready-to-run code snippets helped us bridge gaps in our programming knowledge, particularly in integrating multiple NHANES cycles and performing regression modeling with interaction terms. This allowed us to focus more on interpreting the meaning behind the numbers and less on debugging code.

**Influence on Understanding and Approach**

 AI support shifted our approach from trial-and-error coding toward a more hypothesis-driven analysis. For instance, AI-suggested plots of thyroid hormone levels by ethnicity and gender highlighted patterns we might have overlooked, such as subtle demographic variations in hormone distribution. Similarly, regression output summaries encouraged us to consider the complex interplay between BMI, sleep, and depression on thyroid function rather than isolated effects.

Having AI provide alternate analytical perspectives prompted critical thinking and iterative refinement. We began framing our research question not just as “Are these factors associated with thyroid levels?” but “How might these factors interact to affect thyroid function in diverse populations?” This reflection was only possible because AI presented us with multiple statistical avenues and visualization tools.

**Insights Gained**

 The project underscored the importance of consistent variable naming and clear documentation, especially when working with large, multi-cycle datasets like NHANES. We learned to check for and appropriately handle missing or ambiguous data codes (e.g., “Refused” or “Don’t know”) early to avoid bias in analysis.

Another key insight was the necessity of examining data distribution and testing assumptions before selecting statistical tests. AI’s recommendation of Shapiro–Wilk tests and QQ plots was invaluable in guiding us away from inappropriate parametric tests when normality was violated.

We also appreciated AI as a “second opinion” that exposed us to coding and analytical strategies outside our immediate expertise, reinforcing the value of interdisciplinary collaboration between human insight and AI assistance.

**Challenges Encountered**

 Despite these benefits, AI sometimes generated code that assumed ideal conditions—such as variables existing in all NHANES cycles or data being free of inconsistencies—which led to errors or incomplete merges. This required us to critically inspect AI outputs, validate against source documentation, and manually adjust or extend code.

At times, the AI’s statistical interpretations were generic, lacking biomedical context. For example, while it correctly described p-values and coefficients, it did not address clinical relevance or biological plausibility, which remained our responsibility.

Moreover, AI was limited in managing NHANES-specific nuances like survey weighting and complex sampling design, which are critical for population-level inference. We had to be cautious not to over-rely on AI-generated code without integrating domain knowledge.

**Limitations and Shortcomings**

 AI tools currently lack awareness of domain-specific complexities and the subtle judgment required for public health data analysis. Its outputs, while efficient, sometimes favored simplicity over statistical rigor—defaulting to linear regression without assessing model assumptions or suggesting advanced modeling approaches.

In essence, AI served as an effective assistant, but not a substitute for critical thinking, domain expertise, or manual data vetting. The iterative human-AI partnership was essential: AI accelerated coding and expanded analytical possibilities, but human oversight ensured validity, relevance, and interpretability.
